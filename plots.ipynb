```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result.csv')
df
```

```python
for key, grp in df.groupby(df.columns[0]):
    plt.plot(grp[df.columns[1]], grp[df.columns[2]], label=key)

plt.title('Mining Power Utilization')
plt.xlabel(df.columns[1])
plt.ylabel(df.columns[2])
plt.legend(title=df.columns[0])
plt.show()
```

```python
for key, grp in df.groupby(df.columns[0]):
    plt.plot(grp[df.columns[1]],grp[df.columns[3]], label=key)

plt.title('Fraction Mined by Adversary')
plt.xlabel(df.columns[1])
plt.ylabel(df.columns[3])
plt.legend(title=df.columns[0])
plt.show()
```
